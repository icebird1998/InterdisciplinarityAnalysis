```python
import pandas as pd
from nltk import word_tokenize, ngrams
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from collections import Counter
import nltk
import re
import string

# 下载必要资源
nltk.download('punkt')
nltk.download('stopwords')

# 初始化工具
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
punct_set = set(string.punctuation)
ngram_freq = Counter()

# 参数：你希望统计的 n 值范围，例如 2 到 4
n_values = [1, 2, 3, 4, 5]

# 清理 <sup> 和 <sub>
def clean_html_tags(text):
    text = re.sub(r'<sup>.*?</sup>', '', text, flags=re.DOTALL)
    text = re.sub(r'<sub>.*?</sub>', '', text, flags=re.DOTALL)
    return text

# 主处理流程
for abstract in cite_information['abstract']:
    if pd.notnull(abstract):
        # 清理标签
        clean_text = clean_html_tags(abstract)
        # 分词 + 小写
        tokens = word_tokenize(clean_text.lower())
        # 去停用词、标点、非字母词 + 词干还原
        filtered_tokens = [
            stemmer.stem(word)
            for word in tokens
            if word.isalpha() and word not in stop_words and word not in punct_set
        ]
        # 多个 n 值的 n-gram 提取
        for n in n_values:
            n_grams = ngrams(filtered_tokens, n)
            ngram_freq.update(n_grams)

# 转为 DataFrame
ngram_df = pd.DataFrame(ngram_freq.items(), columns=['ngram', 'frequency'])
ngram_df['ngram_length'] = ngram_df['ngram'].apply(len)
ngram_df = ngram_df.sort_values(by='frequency', ascending=False)

# 可选：把 ngram 元组转成字符串显示
ngram_df['ngram'] = ngram_df['ngram'].apply(lambda x: ' '.join(x))

# 显示前10个高频短语
print(ngram_df.head(10))
```

```python
import pandas as pd
from nltk import word_tokenize, ngrams
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from collections import Counter
import nltk
import re
import string
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 下载资源
nltk.download('punkt')
nltk.download('stopwords')

# 假设 cite_information 已经存在，这里用一个示例 DataFrame 创建占位
# cite_information = pd.DataFrame({'abstract': ['Rare earth elements are important.', 'Coal seam gas and coal bed methane are resources.']})

# 初始化工具
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
punct_set = set(string.punctuation)
ngram_freq = Counter()
n_values = [1, 2, 3]

# 清洗 HTML 标签函数
def clean_html_tags(text):
    text = re.sub(r'<sup>.*?</sup>', '', text, flags=re.DOTALL)
    text = re.sub(r'<sub>.*?</sub>', '', text, flags=re.DOTALL)
    return text

# 主处理逻辑
for abstract in cite_information['abstract']:
    if pd.notnull(abstract):
        clean_text = clean_html_tags(abstract)
        tokens = word_tokenize(clean_text.lower())
        filtered_tokens = [
            stemmer.stem(word)
            for word in tokens
            if word.isalpha() and word not in stop_words and word not in punct_set
        ]
        for n in n_values:
            n_grams = ngrams(filtered_tokens, n)
            ngram_freq.update(n_grams)

# 创建 DataFrame
ngram_df = pd.DataFrame(ngram_freq.items(), columns=['ngram', 'frequency'])
ngram_df['ngram_length'] = ngram_df['ngram'].apply(len)
ngram_df['ngram_str'] = ngram_df['ngram'].apply(lambda x: ' '.join(x))

# 分别绘制 n=1, 2, 3 的词云图
for n in n_values:
    subset = ngram_df[ngram_df['ngram_length'] == n].nlargest(25, 'frequency')
    word_freq = dict(zip(subset['ngram_str'], subset['frequency']))
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
    
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f'{n}-gram Top 25 WordCloud')
    plt.axis('off')
    plt.show()
```

```python
# 保存为3个CSV文件，每个n-gram一个
csv_paths = {}
for n in n_values:
    subset = ngram_df[ngram_df['ngram_length'] == n].nlargest(25, 'frequency')
    path = f'./ngram_top25_{n}gram.csv'
    subset[['ngram_str', 'frequency']].to_csv(path, index=False)
    csv_paths[f'{n}-gram'] = path

csv_paths
```
